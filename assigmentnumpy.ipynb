```python
#NumPy (Numerical Python) is an open-source Python library used for numerical and scientific computing
#Supports 1D, 2D, 3D, and higher-dimensional arrays.
#Uses less memory because all elements have the same data type
#ndim returns the number of dimensions (axes) of the array.
#shape returns the size of the array in each dimension.
#size returns the total number of elements in the array.
#dtype returns the data type of the elements stored in the array.
#type() returns the type of the object.
#Python List	Part of Python's built-in data types
##NumPy Array Part of Python's built-in data types
#shape returns the dimensions of the array
#size Returns the total number of elements
#dtype Returns the data type of the array elements
#type()	Returns the type of the object
#In a 3D NumPy array, the shape is written as (a, b, c).
#a - Number of matrices
#b - Number of rows in each matrix
#c - Number of columns in each row
```
